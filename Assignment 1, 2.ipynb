#### Assignment 1

```r
library(drc)
library(e1071)
data(vinclozolin)

conc <- vinclozolin$conc
effect <- vinclozolin$effect
exper <- vinclozolin$exper

explore <- function(data) {
    cat('max:', max(data), '\n')
    cat('min:', min(data), '\n')
    cat('mean:', mean(data), '\n')
    
    cat('std:', sd(data), '\n')
    cat('skewness:', skewness(data), '\n')
    cat('kurtosis:', kurtosis(data), '\n')
}
```

```r
cat('Summary statistics of Conc', '\n')
explore(conc)
```

```r
cat('Summary statistics of Effect', '\n')
explore(effect)
```

```r
cat('Correlation between Effect and Conc', '\n')
cat(cor(effect, conc))
```

#### Assignment 2

```r
library(ggplot2)
library(repr)
options(repr.plot.width=5, repr.plot.height=3)

ggplot(data = vinclozolin, mapping = aes(x = conc, y = effect)) + 
geom_point(aes(colour = exper), shape = 19) + 
ggtitle("Vinclozolin")
```

#### Assignment 3, 4

As seen from the plot above, the relationship between `conc` and `effect` depends on `exper`. So we split data by `exper`, which is better demonstrated on the plots below:

```r
options(repr.plot.width=10, repr.plot.height=5)
par(mfrow=c(2,3))
for (val in unique(exper)) {
    plot(conc[exper == val], effect[exper == val], pch=19, xlab='Conc', ylab='Effect', main=c('Exper: ', val))
}
```

As seen from the plots above, after separation, the relationship between `conc` and `exper` is much more obvious. So, it's reasonable to have different models for different `exper`s.
